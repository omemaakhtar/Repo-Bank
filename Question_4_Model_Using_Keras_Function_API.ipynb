```python
from keras.models import Model
from keras.layers import Input, Dense, concatenate
from keras.utils import plot_model

# Define inputs
input1 = Input(shape=(32,), name='input1')  # Input Layer 1
input2 = Input(shape=(128,), name='input2')  # Input Layer 2

# Branch for input1
dense1 = Dense(8, activation='relu', name='dense1')(input1)
dense2 = Dense(4, activation='relu', name='dense2')(dense1)

# Branch for input2
dense3 = Dense(32, activation='relu', name='dense3')(input2)
dense4 = Dense(8, activation='relu', name='dense4')(dense3)
dense5 = Dense(4, activation='relu', name='dense5')(dense4)

# Concatenate dense2 and dense5
concatenate1 = concatenate([dense2, dense5], name='concatenate1')

# Further dense layers
dense6 = Dense(2, activation='relu', name='dense6')(concatenate1)
output = Dense(1, activation='sigmoid', name='output')(dense6)

# Create model
model = Model(inputs=[input1, input2], outputs=output)

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Save model plot to a file
plot_model(model, to_file='model.png', show_shapes=True, show_layer_names=True)

# Optionally display the model within the notebook (if using Jupyter)
from IPython.display import Image
Image('model.png')
```
