```python
# Retrive weights from the second hidden Layer
filters,bias = model.layers[1].get_weights()
```

```python
# Normalize filter vlaue to 0-1 so we can visualize them
f_min,f_max = filters.min(),filters.max()
filters = (filters-f_min)/(f_max-f_min)
```

```python
import matplotlib.pyplot as plt
```

```python
n_filters = 6
ix=1
fig=plt.figure(figsize=(15,10))
for i in range(n_filters):
  f=filters[:,:,:,i]
  for j in range(3):
    plt.subplot(n_filters,3,ix)
    plt.imshow(f[:,:,j],cmap='gray')
    ix += 1
plt.show()
```

```python
from keras.applications.vgg16 import VGG16
from matplotlib import pyplot
# load the model
model = VGG16()
# summarize feature map shapes
for i in range(len(model.layers)):
 layer = model.layers[i]
 # check for convolutional layer
 if 'conv' not in layer.name:
  continue
 # summarize output shape
 print(i, layer.name, layer.output.shape)
```

```python
from keras.layers import Lambda, concatenate
from keras import Model
import tensorflow as tf
```

```python
# redefine model to output right after the first hidden layer
model = Model(inputs=model.inputs, outputs=model.layers[1].output)
```

```python
#Input shape to the model is 224 x 224. SO resize input image to this shape.
from keras.preprocessing.image import load_img, img_to_array
import tensorflow as tf
import numpy as np
```

```python
#Input shape to the model is 224 x 224. SO resize input image to this shape.
from keras.preprocessing.image import load_img, img_to_array
img = load_img('elonmusk.jfif', target_size=(224, 224)) #VGG user 224 as input

# convert the image to an array
img = img_to_array(img)
# expand dimensions to match the shape of model input
img = np.expand_dims(img, axis=0)
# Generate feature output by predicting on the input image
feature_output = model.predict(img)
```

```python
#calculating features_map

fig = pyplot.figure(figsize=(20,15))
for i in range(1,feature_output.shape[3]+1):

    pyplot.subplot(8,8,i)
    pyplot.imshow(feature_output[0,:,:,i-1] , cmap='gray')

pyplot.show()
```
